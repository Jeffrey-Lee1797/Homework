# Part 1: Canids Continued

```python
# open the file (that was cleaned in the shell)
infile = "clean_canid.csv"
with open(infile, 'r') as ff:
    canid_recs = ff.readlines()[1:]
```

```python
def sort_name(data):
    return(data.split(",")[5])
    
    
another_csv = sorted(canid_recs, key = sort_name)   

output_file = open("sorted_canids.csv", "w")
for entry in another_csv:
    output_file.write(entry)
    
output_file.close()
```

## First function

```python
def canid_dict(csv):
    from collections import defaultdict
    
    with open(csv, "r") as aa:
        csv_data = aa.readlines()
    
    species_ranges = defaultdict(list)

    for line in csv_data:
        line = line.replace('"', '')
        species = line.split(",")[5]
        if len(species.split(" ")) > 1:
            minage = float(line.split(',')[10])
            maxage = float(line.split(',')[11])
            meanage = (minage + maxage) / 2
    
            species_ranges[species].append(meanage) 
    return species_ranges
```

```python
new_dict = canid_dict("sorted_canids.csv")
new_dict
```

## Second function

```python
def canid_summary(dictionary, outputfile):
    output = open(str(outputfile), "w")
    for key in sorted(dictionary.keys()): 
        ages = dictionary[key]
        maxage = max(ages)
        minage = min(ages)
        genus = key.split(" ")[0] 
        species =key.replace(" ", "_")  
        outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
        output.write(outline)
    output.close()
```

```python
canid_summary(new_dict, "canid-ranges.csv")
```

## Third function 

```python
with open("canid-ranges.csv", "r") as jj:
    canid_ranges = jj.readlines()
rangedict = {}
for line in canid_ranges:
    element = line.split(",")
    species = element[1]
    fossilrange = float(element[3])-float(element[2])
    rangedict[species] = fossilrange
```

```python
rangedict
```

```python
def longest_range(dictionary):
    longest_species = max(dictionary, key=dictionary.get)
    return longest_species
```

```python
longest_range(rangedict)
```

```python
infile = "Ursidae_data_cleaned.csv"
with open(infile, 'r') as rr:
    Ursidae_recs = rr.readlines()[1:]
```

## First function

```python
def Ursidae_dict(csv):
    from collections import defaultdict
    
    with open(csv, "r") as bb:
        Ursidae_data = bb.readlines()
        
    species_ranges = defaultdict(list)

    for line in Ursidae_data:
        line = line.replace('"', '')
        species = line.split(",")[5]
        if len(species.split(" ")) > 1:
            minage = float(line.split(',')[14])
            maxage = float(line.split(',')[15])
            meanage = (minage + maxage) / 2
    
            species_ranges[species].append(meanage) 
    return species_ranges
```

```python
Ursidae_new_dict = Ursidae_dict("sorted_Ursidae.csv")
Ursidae_new_dict
```

##  Second function

```python
def Ursidae_summary(dictionary, outputfile):
    output = open(str(outputfile), "w")
    for key in sorted(dictionary.keys()): 
        ages = dictionary[key]
        maxage = max(ages)
        minage = min(ages)
        genus = key.split(" ")[0] 
        species =key.replace(" ", "_")  
        outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
        output.write(outline)
    output.close()
```

```python
Ursidae_summary(Ursidae_new_dict, "Ursidae-ranges.csv")
```
